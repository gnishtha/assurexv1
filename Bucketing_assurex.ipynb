```
IF WE KEEP A CHECK ON WHAT IS VISITED ALREADY AND DO NOT CONSIDER THEM AGAIN:
```

```
res={}
prev=[]
temp=[]
for token_1 in x:
  prev.append(token_1)
  for token_2 in x:
    if token_2 not in prev:
      #print(fuzz.WRatio(token_1 , token_2))    
      if fuzz.WRatio(token_1 , token_2)>80:
        #print("hiiiiiii", token_2)
        temp.append(token_2)
        prev.append(token_2)

  res[token_1]=temp
  temp=[]
```

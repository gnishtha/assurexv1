```

import pandas as pd
import numpy as np
from fancyimpute import KNN
from sklearn.preprocessing import OrdinalEncoder
```

```
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/example2 - Product Mapping.csv')
#df.head()
dfunk=df[df['lobdesc']!='Unknown']
#df.describe
```

```
gk = dfunk.groupby('lobraw') 
gk.head()  
```

```
def findingUtterances(temp):
  temp_dict={}
  flag=True
  l=[]
  count=0
  #print(len(temp))
  for i in range(len(temp)):
    #print(temp.AGLOBDesc.values[i])
    if count!=len(temp) and temp.AGLOBDesc.values[i] not in l:
      #print("nis")
      temp_dict[temp.AGLOBDesc.values[i]]=1
      l.append(temp.AGLOBDesc.values[i])
      #print(count)
      count+=1
    elif temp.AGLOBDesc.values[i] in l:
      temp_dict[temp.AGLOBDesc.values[i]]=temp_dict[temp.AGLOBDesc.values[i]]+1
  return temp_dict
  
```

```
emp = gk.get_group('FMLA')
emp
print(findingUtterances(emp))
```

```
l={}
#lx=[]
for i in range(len(dfunk)):
  if dfunk.lobraw.values[i] not in l.keys():
    temp = gk.get_group(dfunk.lobraw.values[i])
    res = findingUtterances(temp)
    l[dfunk.lobraw.values[i]]=res
```

```
final={}
for k,v in l.items():
  max_key = max(v, key=v.get)
  final[k]=max_key
```

```
dfx = {}
dfpred=df[df['lobdesc']=='Unknown']
for i in range(len(dfpred)):
  try:
    dfx[dfpred.lobraw.values[i]]=final[dfpred.lobraw.values[i]]
  except:
    print(dfpred.lobraw.values[i])
```
